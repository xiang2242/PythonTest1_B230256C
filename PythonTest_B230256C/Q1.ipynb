```python
import random

# List of words for the game
words = ["Apple", "Banana", " Cherries", " Dragon Fruit", "Guava", "Jambulan", "Watermelon"]

# Randomly select a word from the list
word_to_guess = random.choice(words).upper()

# Initialize an empty list to represent the hidden word with underscores
hidden_word = ["_" for _ in word_to_guess]

# Set the number of allowed incorrect guesses
max_incorrect_guesses = 6
incorrect_guesses = 0

# List to keep track of guessed letters
guessed_letters = []

# Hangman stages for visualization
hangman_stages = [
    """
     ------------
     |          |
     |          |
     |          |
     |          |
     |          |
     ------------
    """,
    """
     ------------
     |          |
     |    O     |
     |          |   
     |          |
     |          |
     ------------
    """,
    """
     ------------
     |          |
     |    O     |
     |    |     |
     |          |
     |          |
     ------------
    """,
    """
     ------------
     |          |
     |    O     |
     |   /|     |
     |          |
     |          |
     ------------
    """,
    """
     ------------
     |          |
     |    O     |
     |   /|\    |
     |          |
     |          |
     ------------
    """,
    """
     ------------
     |          |
     |    O     |
     |   /|\    |
     |   /      |
     |          |
     ------------
    """,
    """
     ------------
     |          |
     |    O     |
     |   /|\    |
     |   / \    |
     |          |
     ------------
    """
]

# Function to display the current state of the game
def display_game_state():
    print(hangman_stages[incorrect_guesses])
    print(f" {' '.join(hidden_word)}")
    print(f"Incorrect guesses: {incorrect_guesses}")
    print()

# Game loop
while incorrect_guesses < max_incorrect_guesses and "_" in hidden_word:
    display_game_state()
    
    guess = input("Guess a letter: ").upper()

    if guess in guessed_letters:
        print("You already guessed that letter.")
        continue
    
    guessed_letters.append(guess)
    
    if guess in word_to_guess:
        for i, letter in enumerate(word_to_guess):
            if letter == guess:
                hidden_word[i] = guess
        print(f"Good guess! {guess} is in the word.\n")
    else:
        incorrect_guesses += 1
        print(f"Incorrect! {guess} is not in the word.\n")
    
# End of game
display_game_state()

if "_" not in hidden_word:
    print(f"Congratulations! You've guessed the word: {word_to_guess}")
else:
    print(f"Sorry, you've run out of guesses. The word was: {word_to_guess}")
```
